```python
# 2609
num1, num2 = map(int, input().split())

# 최대공약수
if num1 > num2 :
    for i in range(num2, 0, -1) :
        if (num1 % i == 0) and (num2 % i == 0) :
            gcd = i
            print(gcd)
            break
else :
    for i in range(num1, 0, -1) :
        if (num1 % i == 0) and (num2 % i == 0) :
            gcd = i
            print(gcd)
            break

# 최소공배수(최대공약수 활용)
lcm = int((num1 * num2) / gcd)
print(lcm)
```

```python
# 11650
import sys

input_num = int(sys.stdin.readline().rstrip())
number_arr = []

for i in range(input_num) :
  num1, num2 = map(int, sys.stdin.readline().rstrip().split())
  binding = [num1, num2]

  number_arr.append(binding)

number_arr.sort()

for i in range(0, len(number_arr)) :
  print(int(number_arr[i][0]), int(number_arr[i][1]))
```

```python
# 2839
N = int(input())

max_5kg = int(N / 5)

if N % 5 == 0 :
  print(max_5kg)
else :
  if (N % 5) % 3 == 0 :
    print(int(N / 5 + (N % 5) / 3))
  else :
    chk = 0
    if N % 3 == 0 :
      chk = 1
      result1 = int(N / 3)

      for i in range(max_5kg - 1, 0, -1) :
        if (N - i * 5) % 3 == 0 :
          result2 = int(i + (N - i * 5) / 3)
          chk = 2
          break

    for i in range(max_5kg - 1, 0, -1) :
      if (N - i * 5) % 3 == 0 :
         result2 = int(i + (N - i * 5) / 3)
         chk = 3
         break

    if chk == 1 :
      print(result1)

    elif chk == 2 :
      print(min(result1, result2))

    elif chk == 3 :
      print(result2)

    else :
      print(-1)
```

```python
# 1966
import sys

cases = int(input())

for i in range(cases) : 
  pages, page_idx = map(int, input().split())
  prior_list = list(map(int, sys.stdin.readline().strip().split()))
  page_chk = prior_list[page_idx] + 0.5
  prior_list[page_idx] = prior_list[page_idx] + 0.5
  cnt = 1

  while 1 :
    if len(prior_list) < 2 :
      print(cnt)
      break

    else :
      if int(prior_list[0]) == int(max(prior_list)) :
        if str(prior_list[0]) == str(page_chk) :
          print(cnt)
          break
        else :
          del prior_list[0]
          cnt += 1
      else :
        prior_list.append(prior_list[0])
        del prior_list[0]
```

```python
# 10828
import sys

N = int(input())
stack = []

for i in range(N) :
  text = sys.stdin.readline().strip()

  if "push" in text :
    replacedText = text.replace("push ", "")
    stack.insert(0, int(replacedText))
  elif text == "pop" :
    if not stack :
      print(-1)
    else :
      print(stack.pop(0))
  elif text == "size" :
    print(len(stack))
  elif text == "empty" :
    if not stack :
      print(1)
    else :
      print(0)
  elif text == "top" :
    if not stack :
      print("-1")
    else :
      print(stack[0])
```
