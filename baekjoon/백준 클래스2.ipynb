```python
# 1654 랜선 자르기
need_ea, need_len = map(int, input().split())
cable = [int(input()) for i in range(need_ea)]

sum_lst = []
lst = []

left = 0
right = max(cable)

while left <= right :
    mid = (right + left) // 2
    sum = 0

    if mid == 0 :
        mid = 1

    for sample in cable :
        sum += (sample // mid)
    if sum >= need_len :
        sum_lst.append(mid)
        left = mid + 1
    else :
        right = mid - 1

sum_lst.sort()

print(sum_lst[-1])
```

```python
# 18111 마인크래프트
import sys
from collections import deque

N, M, B = map(int, sys.stdin.readline().strip().split())
lst = []
result = []

for i in range(N) :
    lst.append(sys.stdin.readline().strip().split())

lst = sum(lst, [])
lst = list(map(int, lst))

high = max(lst)
low = min(lst)

for i in range(low, high + 1) :
    pockets = B
    cnt = 0
    for sample in lst :
        if i > sample :
            compensate = i - sample
            cnt += 1 * compensate
            pockets -= 1 * compensate
        elif i < sample :
            compensate = sample - i
            cnt += 2 * compensate
            pockets += 1 * compensate

    if pockets >= 0 :
        binding = [cnt, i]
        result.append(binding)

result = sorted(result, key = lambda x : (x[0], -x[1]))

print(*result[0])
```
